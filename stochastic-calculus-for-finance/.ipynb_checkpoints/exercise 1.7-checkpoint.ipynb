**Exercise 1.7** (Hedging a long position-multiple periods). Consider a bank that has a long position in the lookback option of Example 1.2.3. The bank intends to hold this option until expiration and receive the payoff $V_3$. At time zero, the bank has capital $V_0 = 1.376$ tied up in the option and wants to earn the interest rate of $25$\% on this capital until time three(i.e. without investing any more money, and regardless of how the coin tossing turns out, the bank wants to have

$$\left(\frac{5}{4}\right)^3 (1.376) = 2.6875$$

at time three, after collecting the payoff from the lookback option at time three.) Specify how the bank's trader should invest in the stock and the money market account to accomplish this.

*Solution.*

---
The exercise asks how to invest in the stock and the money market account so that our final portfolio value at the end of time three is,

\begin{align*}
\left(\frac{5}{4}\right)^3(1.376) = (1+r)^3 V_0
\end{align*}

---

This is similar to the previous example. The bank's trader would construct a hedging portfolio who payoff is exactly opposite to the option payoff. More precisely, at **time one**, we would like to have

\begin{align*}
\Delta_0 S_1 + (1+r)(X_0 - \Delta_0 S_0) &= -V_1 
\end{align*}

So,

\begin{align*}
8\Delta_0 + 1.25(X_0 - 4\Delta_0) &= -2.24\\
2\Delta_0 + 1.25(X_0 - 4\Delta_0) &= -1.20\\
\end{align*}

That is,

\begin{align*}
3\Delta_0 + 1.25X_0 &= -2.24\\
-3\Delta_0 + 1.25X_0&= -1.20
\end{align*}

```julia
Δ₀ = (-2.24 + 1.20)/6
Δ₀
```

```julia
X₀ = (-2.24-1.20)/2.5
X₀
```

Thus, at time zero, the agent should short-sell 0.173333 units of the stock for an income of 0.69333 dollars, borrow 0.682 dollars from a money market account and keep the sum total 1.376 dollars in a separate money market account. The hedging portfolio $(-0.173333,-0.682)$ cancels out any gains from the option contract at time one.

```julia
s0, s1_h, s1_t = 4, 8, 2;
delta0 = -0.173333;
x0 = -1.37666;

s1_h * delta0 + (1 + r)*(x0 - delta0 * s0),
s1_t * delta0 + (1 + r)*(x0 - delta0 * s0)
```

At **time two** we would like to have:

\begin{align*}
\Delta_1 S_2 + (1+r)(X_1 - \Delta_1 S_1) = -V_2
\end{align*}

So, it gives two pairs of equations:

\begin{align*}
\Delta_1(H) S_2(HH) + (1+r)(X_1(H) - \Delta_1(H) S_1(H)) &= -V_2(HH) \\
\Delta_1(H) S_2(HT) + (1+r)(X_1(H) - \Delta_1(H) S_1(H)) &= -V_2(HT) \\
\Delta_1(T) S_2(TH) + (1+r)(X_1(T) - \Delta_1(T) S_1(T)) &= -V_2(TH) \\
\Delta_1(T) S_2(TT) + (1+r)(X_1(T) - \Delta_1(T) S_1(T)) &= -V_2(TT) 
\end{align*}


\begin{align*}
16\Delta_1(H) + 1.25(X_1(H) - 8\Delta_1(H)) = -3.20 \\
4\Delta_1(H) + 1.25(X_1(H) - 8\Delta_1(H)) = -2.40 \\
4\Delta_1(T) + 1.25(X_1(T) - 2\Delta_1(T)) = -0.80\\
\Delta_1(T) + 1.25(X_1(T) - 2\Delta_1(T)) = -2.20
\end{align*}

\begin{align*}
6\Delta_1(H) + 1.25X_1(H) &= -3.20 \\
-6\Delta_1(H) + 1.25X_1(H) &= -2.40 \\
1.5\Delta_1(T) + 1.25X_1(T) &= -0.80\\
-1.5\Delta_1(T) + 1.25X_1(T) &= -2.20
\end{align*}

So, $\Delta_1(H) = -0.06667$, $X_1(H)=-2.23999$, $\Delta_1(T) = 0.4667$ and $X_1(T) = -1.20$

If the first coin toss lands heads, then at time one the agent has a portfolio valued at $X_1(H)$ and she re-balances her hedge, she must short-sell 0.06667 units of the stock for an income of 0.53336 dollars and borrow 1.706663 dollars from the money market account. So, she updates her hedge to $(-0.06667,-1.706663)$. This cancels out any gains from the option contract at time two.

```julia
Δ₁ = -0.066667
S₁ = 8
X₁ = -2.239999
r = 0.25
```

```julia
Δ₁*S₁
```

```julia
X₁ - Δ₁* S₁
```

```julia
16*Δ₁ + (1+r)*(X₁ - Δ₁* S₁),4*Δ₁ + (1+r)*(X₁ - Δ₁* S₁)
```

If the first coin toss lands tails, then the agent has a portfolio valued at $X_1(T)$ and she re-adjusts her hedge, she must buy 0.46667 units of the stock for  and borrow 2.13334 dollars from the money market account. So, she updates her hedge to $(0.4667,-2.13334)$. This cancels out any gains from the option contract at time two.

```julia
Δ₁ = 0.466667
S₁ = 2
X₁ = -1.20
r = 0.25
```

```julia
Δ₁*S₁
```

```julia
X₁ - Δ₁* S₁
```

```julia
4*Δ₁ + (1+r)*(X₁ - Δ₁* S₁),1*Δ₁ + (1+r)*(X₁ - Δ₁* S₁)
```

At **time three**, we would like to have:


\begin{align*}
\Delta_2 S_3 + (1+r)(X_2 - \Delta_2 S_2) = -V_3
\end{align*}

This give us:

\begin{align*}
\Delta_2(HH) S_3(HHH) + (1+r)(X_2(HH) - \Delta_2(HH) S_2(HH)) &= -V_3(HHH)\\
\Delta_2(HH) S_3(HHT) + (1+r)(X_2(HH) - \Delta_2(HH) S_2(HH)) &= -V_3(HHT)\\
\Delta_2(HT) S_3(HTH) + (1+r)(X_2(HT) - \Delta_2(HT) S_2(HT)) &= -V_3(HTH)\\
\Delta_2(HT) S_3(HTT) + (1+r)(X_2(HT) - \Delta_2(HT) S_2(HT)) &= -V_3(HTT)\\
\Delta_2(TH) S_3(THH) + (1+r)(X_2(TH) - \Delta_2(TH) S_2(TH)) &= -V_3(THH)\\
\Delta_2(TH) S_3(THT) + (1+r)(X_2(TH) - \Delta_2(TH) S_2(TH)) &= -V_3(THT)\\
\Delta_2(TT) S_3(TTH) + (1+r)(X_2(TT) - \Delta_2(TT) S_2(TT)) &= -V_3(TTH)\\
\Delta_2(TT) S_3(TTT) + (1+r)(X_2(TT) - \Delta_2(TT) S_2(TT)) &= -V_3(TTT)
\end{align*}

\begin{align*}
32 \Delta_2(HH) + 1.25(X_2(HH) - 16\Delta_2(HH)) &= 0\\
8 \Delta_2(HH) + 1.25(X_2(HH) - 16\Delta_2(HH)) &= -8\\
8\Delta_2(HT) + 1.25(X_2(HT) - 4\Delta_2(HT)) &= 0\\
2\Delta_2(HT) + 1.25(X_2(HT) - 4\Delta_2(HT)) &= -6\\
8\Delta_2(TH)  + 1.25(X_2(TH) - 4\Delta_2(TH)) &= 0\\
2\Delta_2(TH)  + 1.25(X_2(TH) - 4\Delta_2(TH)) &= -2\\
2\Delta_2(TT) + 1.25(X_2(TT) - \Delta_2(TT)) &= -2\\
0.5\Delta_2(TT) + 1.25(X_2(TT) - \Delta_2(TT)) &= -3.5
\end{align*}

Thus, at time two, the agent updates her delta hedges according to:

```julia
delta_hh,delta_ht,delta_th,delta_tt = (3.20 * 1.25)/(32-16*1.25), (2.40 * 1.25)/(8-4*1.25), (1.25*0.80)/(8-4*1.25), (-2 + 1.25*2.20)/(2-1.25)
```

And the amounts she should invest in the money market account are:

```julia
x2_hh, x2_ht, x2_th, x2_tt = -3.20, -2.40, -0.80, -2.20
s2_hh, s2_ht, s2_th, s2_tt = 16,4,4,1
(x2_hh - s2_hh * delta_hh), (x2_ht - s2_ht * delta_ht), (x2_th - s2_th * delta_th), (x2_tt - s2_tt * delta_tt)
```

The hedge portfolio offsets the performance of the option at time three.

```julia
s3_hhh, s3_hht, s3_hth, s3_htt  = 32, 8, 8, 2;
s3_thh, s3_tht, s3_tth, s3_ttt   = 8, 2, 2, 0.5;
```

```julia
s3_hhh * delta_hh + (1+r)*(x2_hh - s2_hh * delta_hh),
s3_hht * delta_hh + (1+r)*(x2_hh - s2_hh * delta_hh),
s3_hth * delta_ht + (1+r)*(x2_ht - s2_ht * delta_ht),
s3_htt * delta_ht + (1+r)*(x2_ht - s2_ht * delta_ht),
s3_thh * delta_th + (1+r)*(x2_th - s2_th * delta_th),
s3_tht * delta_th + (1+r)*(x2_th - s2_th * delta_th),
s3_tth * delta_tt + (1+r)*(x2_tt - s2_tt * delta_tt),
s3_ttt * delta_tt + (1+r)*(x2_tt - s2_tt * delta_tt)
```

At at end of each time period, the agent only rebalances her hedge-ratio (stocks:cash), that is her replication portfolio wealth $X_t$ remains unchanged. So, the $1.376$ dollars she invested in a separate MMA, would grow to $(5/4)^3(1.376)$.
