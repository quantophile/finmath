**Exercise 1.6** (Hedging a long position-one period). Consider a bank that has a long position in the European call written on the stock price in figure 1.1.2. The call expires at time one and has strike price $K=5$. In section 1.1, we determined the time-zero price of this call to be $V_0 = 1.20$. At time zero, the bank owns this option, which ties up capital $V_0 = 1.20$. The bank wants to earn the interest rate $25$\% on this capital until time one i.e. without investing any more money, regardless of how the coin tossing turns out, the bank wants to have

\begin{align*}
\frac{5}{4} \cdot 1.20 = 1.50
\end{align*}

at time one, after collecting the payoff from the option (if any) at time one). Specify how the bank's trader should invest in the stock and money markets to accomplish this.

*Solution.*

The bank's trader should set up a replicating portfolio whose payoff is the exact opposite of the option's payoff. More precisely, we would like to solve the equation:

\begin{align*}
\Delta_0 S_1 + (1+r)(X_0 - \Delta_0 S_0) = -(S_1 - K)^{+}
\end{align*}

Then,

\begin{align*}
8 \Delta_0 + 1.25(X_0 - 4\Delta_0) &= -3 \\
2 \Delta_0 + 1.25(X_0 - 4\Delta_0) &= 0
\end{align*}

Simplifying,
\begin{align*}
3 \Delta_0 + 1.25X_0 &= -3 \\
-3 \Delta_0 + 1.25X_0 &= 0
\end{align*}

So, $\Delta_0 = -0.5$ and $X_0 = -1.20$. 

Thus, the trader should short-sell $\Delta_0 S_0 = 2$ dollars worth of stock, invest $0.80$ into a money-market account and the remaining $1.20$ into a separate money market account. The losses on the hedging portfolio will square off any gains on the European call option, and the $1.20$ in the separate bank account will grow to $1.50$ at time 1.
